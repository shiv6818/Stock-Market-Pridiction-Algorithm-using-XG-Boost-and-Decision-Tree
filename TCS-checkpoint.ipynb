```python
splitted = df['Date'].str.split('/', expand=True)

df['day'] = splitted[0].astype('int')
df['month'] = splitted[1].astype('int')
df['year'] = splitted[2].astype('int')

df.head(2476)
```

```python
df['is_quarter_end'] = np.where(df['month']%3==0,1,0)
df.head(2476)
```

```python
data_grouped = df.groupby('year').mean()
plt.subplots(figsize=(20,10))

for i, col in enumerate(['Open Price', 'High Price', 'Low Price', 'Close Price']):
    plt.subplot(2,2,i+1)
    data_grouped[col].plot.bar()
plt.show()
```

```python
df.groupby('is_quarter_end').mean()
```

```python
df['Open-Close'] = df['Open Price'] - df['Close Price']
df['low-high'] = df['Low Price'] - df['High Price']
df['target'] = np.where(df['Close Price'].shift(-1) > df['Close Price'], 1, 0)
```

```python
plt.pie(df['target'].value_counts().values,
        labels=[0, 1], autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

sb.heatmap(df.corr() > 0.9, annot=True, cbar=False)
plt.show()
```

```python
features = df[['Open-Close', 'low-high', 'is_quarter_end']]
target = df['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_valid, Y_train, Y_valid = train_test_split(
features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_valid.shape)
```

```python
models = [LogisticRegression(), SVC(
kernel='poly', probability=True), XGBClassifier()]

for i in range(3):
    models[i].fit(X_train, Y_train)

print(f'{models[i]} : ')
print('Training Accuracy : ', metrics.roc_auc_score(
Y_train, models[i].predict_proba(X_train)[:,1]))
print('Validation Accuracy : ', metrics.roc_auc_score(
Y_valid, models[i].predict_proba(X_valid)[:,1]))
print()
```

```python
cm = confusion_matrix(Y_valid, models[0].predict(X_valid))
sb.heatmap(cm, annot=True, cmap='Blues', fmt='d')
plt.xlabel('Predicted Lable')
plt.ylabel('True Lable')
plt.show()

```

```python
cm = confusion_matrix(Y_valid, models[0].predict(X_valid))

sb.heatmap(cm, annot=True, cmap='Blues', fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

classification_report = metrics.classification_report(Y_valid, models[0].predict(X_valid))
print(classification_report)
```

```python
import pandas as pd

df = pd.read_csv('TCS2.csv')

X = df[['Open Price', 'High Price', 'Low Price', 'WAP', 'No.of Shares']]
y = df['Close Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2022)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LinearRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Suare squared:", r2)

new_data = pd.read_csv('TCS2.csv', nrows=2, usecols=['Open Price', 'High Price', 'Low Price', 'WAP', 'No.of Shares'])
new_data_scaled = scaler.transform(new_data)

predicted_prices = model.predict(new_data_scaled)
print(predicted_prices)
```
